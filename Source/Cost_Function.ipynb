```python
import numpy as np
X = np.array([1,2,3])
Y = np.array([1,2,3])

def cost_func(W, X, Y):
  result = 0
  for i in range(len(X)):
    result += (W*X[i] - Y[i])**2
  return result/len(X)

for feed_W in np.linspace(-3, 5, num=15):
  curr_cost = cost_func(feed_W, X, Y)
  print("{:6.3f} | {:10.5f}".format(feed_W, curr_cost))
```

```python
import tensorflow as tf
X = np.array([1,2,3])
Y = np.array([1,2,3])

def cost_func(W, X, Y):
  y_hat = W * X
  return tf.reduce_mean(y_hat - Y)

for feed_W in tf.linspace(-3, 5, num=15):
  curr_cost = cost_func(feed_W, X, Y)
  print("{:6.3f} | {:10.5f}".format(feed_W, curr_cost))
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

X = np.array([1,2,3])
Y = np.array([1,2,3])

X = [1,2,3,4]
Y = [1,2,3,4]
W = tf.Variable(tf.random.normal([1], -0.2, 0.2))

print('{:5s} | {:11s} | {:10s}'.format('step', 'cost', 'W'))

steps = []
costs = []
ws = []

for step in range(301):
  Y_hat = W * X
  cost = tf.reduce_mean(tf.square(Y_hat - Y))

  alpha = 0.01
  gradient = tf.reduce_mean(tf.multiply(tf.multiply(W, X) - Y, X))
  descent = W - tf.multiply(alpha, gradient)
  W.assign(descent)
  
  if step % 10 == 0:
    print('{:5} | {:11.4f} | {:10.6f}'.format(step, cost.numpy(), W.numpy()[0]))
    ws.append(W.numpy())
    costs.append(cost.numpy())
    steps.append(step)

plt.plot(ws, costs)
plt.title('Cost Function')
plt.xlabel('W')
plt.ylabel('costs')
plt.show()
```
