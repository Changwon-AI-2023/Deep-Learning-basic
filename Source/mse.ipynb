##평균제곱오차(Mean Square Error) 구하기
####mse를 최소화하는 m, b를 구해 보기 바랍니다.

```python
data = [(1, 3), (2, 5)]
m = -1
b = 5
```

```python
def get_mse(data, m, b):
  n = len(data)
  squared_error = 0
  for x, y in data:
    # find predicted y
    y_hat = m*x+b
    squared_error += (y - y_hat)**2

  # get average squared difference
  mse = squared_error / n
  return mse 
```

```python
print(get_mse(data, m, b))
```

```python
import numpy as np
import matplotlib.pyplot as plt
m = -1
b = 5
x = np.array([1, 2])
y = np.array([3, 5])
y_hat = x * m + b
```

```python
plt.plot(x, y, '.')
plt.plot(x, y_hat, '-')
plt.show()

print("Loss:", np.sum((y - y_hat)**2)/len(x))
```
